# autofragment: XYZ (single) example

This notebook runs the **single** workflow on an XYZ file and writes a JSON fragmentation output.

Input: `examples/water_64.xyz`
Output: `examples/output/water_64.fragments.json`

```python
from __future__ import annotations

import json
import subprocess
from pathlib import Path

repo_root = Path.cwd()
examples_dir = repo_root / "examples"
input_xyz = examples_dir / "water_64.xyz"
output_dir = examples_dir / "output"
output_dir.mkdir(parents=True, exist_ok=True)
output_json = output_dir / "water_64.fragments.json"

input_xyz, output_json
```

```python
# Run the CLI subcommand: autofragment single
# Choose how many fragments you want with --n-fragments
cmd = [
    "autofragment",
    "single",
    "--input",
    str(input_xyz),
    "--output",
    str(output_json),
    "--n-fragments",
    "4",
]

print("Running:", " ".join(cmd))
subprocess.run(cmd, check=True)

output_json
```

```python
# Load and inspect
data = json.loads(output_json.read_text())
fragments = data.get("chemical_system", {}).get("fragments", [])

print("version:", data.get("version"))
print("partitioning:", data.get("partitioning"))
print("n_fragments_written:", len(fragments))

# Show the first fragment IDs
print([f.get("id") for f in fragments[:10]])
```
