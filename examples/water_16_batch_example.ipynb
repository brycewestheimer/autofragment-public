# autofragment: XYZ (batch) example

This notebook runs the **batch** workflow to fragment a directory of XYZ files with **consistent labeling** using a reference structure.

Reference: `examples/water_16/water16.000.xyz`
Inputs: all `*.xyz` under `examples/water_16/`
Outputs: `examples/water_16_partitioned/`

```python
from __future__ import annotations

import json
import subprocess
from pathlib import Path

repo_root = Path.cwd()
examples_dir = repo_root / "examples"
input_dir = examples_dir / "water_16"
reference_xyz = input_dir / "water16.000.xyz"
output_dir = examples_dir / "water_16_partitioned"
output_dir.mkdir(parents=True, exist_ok=True)

reference_xyz, input_dir, output_dir
```

```python
# Run the CLI subcommand: autofragment batch
# Choose --n-fragments to control how many fragments you want.
cmd = [
    "autofragment",
    "batch",
    "--reference",
    str(reference_xyz),
    "--input-dir",
    str(input_dir),
    "--output-dir",
    str(output_dir),
    "--n-fragments",
    "4",
]

print("Running:", " ".join(cmd))
subprocess.run(cmd, check=True)

sorted(output_dir.glob('*.json'))[:5]
```

```python
# Load one output file and verify the flat fragments list exists
outputs = sorted(output_dir.glob('*.json'))
if not outputs:
    raise RuntimeError(f"No JSON outputs found in {output_dir}.")

sample_out = outputs[0]
data = json.loads(sample_out.read_text())
fragments = data.get("chemical_system", {}).get("fragments", [])

print("sample_output:", sample_out.name)
print("version:", data.get("version"))
print("partitioning:", data.get("partitioning"))
print("n_fragments_written:", len(fragments))
```
