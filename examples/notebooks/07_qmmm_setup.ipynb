# QM/MM Setup
Defining QM and MM regions for a hybrid calculation.

```python
import sys
sys.path.append('../../src')
from autofragment.partitioners.qmmm import QMMMPartitioner, DistanceSelection
from autofragment.io import read_xyz
import numpy as np
```

```python
# Use Water system for demo
xyz_file = "../data/water_64.xyz"
molecules = read_xyz(xyz_file)

# Select center of geometry
all_coords = np.vstack([m[0].coords for m in molecules])
center = np.mean(all_coords, axis=0)

# Select QM region within 4 Angstroms of center
qm_sel = DistanceSelection(center=center, radius=4.0)

partitioner = QMMMPartitioner(qm_selection=qm_sel, buffer_radius=3.0)
result = partitioner.partition(molecules)

print(f"QM Atoms: {result.n_qm_atoms}")
print(f"Link Atoms: {result.n_link_atoms}")
print(f"MM Atoms: {result.n_mm_atoms}")
```
