# AutoFragment Quickstart
This notebook introduces the core concepts of molecular fragmentation.

```python
import sys
sys.path.append('../../src')
import autofragment as af
from autofragment import MolecularPartitioner
from autofragment.io import read_xyz, write_gamess_fmo
import os
print(f"AutoFragment version: {af.__version__}")
```

## Load Aspirin
We load the aspirin molecule. We treat each atom as a separate unit (atoms_per_molecule=1) 
to demonstrate clustering on a single molecule.

```python
xyz_file = "../data/aspirin.xyz"
system = read_xyz(xyz_file, atoms_per_molecule=1) 
print(f"Loaded aspirin as {len(system)} atomic units")
```

```python
# Partition into 3 fragments
partitioner = MolecularPartitioner(n_fragments=3, method="kmeans")
result = partitioner.partition(system)

print(f"Created {result.n_fragments} fragments.")
for i, frag in enumerate(result.fragments):
    print(f"Fragment {i+1}: {frag.n_atoms} atoms: {frag.symbols}")
```

```python
# Write FMO input
write_gamess_fmo(result.fragments, "aspirin_fmo.inp", method="RHF", basis="6-31G*")
print("Written aspirin_fmo.inp")
if os.path.exists("aspirin_fmo.inp"): os.remove("aspirin_fmo.inp")
```
