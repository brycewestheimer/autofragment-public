# Zeolite Fragmentation
Partitioning a periodic system (Sodalite) using UnitCellPartitioner.

```python
import sys
sys.path.append('../../src')
# Import from geometric partitioners
from autofragment.partitioners.geometric import UnitCellPartitioner
from autofragment.io import read_cif
```

```python
cif_file = "../data/sodalite.cif"
system = read_cif(cif_file)

# Partition 2x2x2 grid
# Note: read_cif returns a ChemicalSystem with lattice info
partitioner = UnitCellPartitioner(grid_shape=(2, 2, 2))
# partition expects ChemicalSystem for matsci partitioners usually, 
# but signature might accept molecules. MatSciPartitioner handles it.
tree = partitioner.partition(system)

print(f"Created {len(tree.fragments)} fragments from unit cell grid.")
print(f"Fragment 0 atoms: {tree.fragments[0].n_atoms}")
```
