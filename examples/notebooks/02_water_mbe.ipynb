# Water Cluster MBE
Fragmentation of a water cluster for Many-Body Expansion (MBE) calculation.

```python
import sys
sys.path.append('../../src')
from autofragment import MolecularPartitioner
from autofragment.io import read_xyz, write_gamess_fmo
import os
```

## Load Water Cluster
We load a 64-water cluster. `read_xyz` defaults to 3 atoms per molecule (O-H-H).

```python
xyz_file = "../data/water_64.xyz"
# Ensure file exists or download it if needed
if not os.path.exists(xyz_file):
    # Retrieve if missing (should exist from previous steps)
    print("Warning: water_64.xyz not found, checking paths.")

molecules = read_xyz(xyz_file)
print(f"Loaded {len(molecules)} water molecules.")
```

## Partition
Group 64 waters into 4 fragments (16 waters each) for FMO.

```python
partitioner = MolecularPartitioner(n_fragments=4, method="kmeans_constrained")
# kmeans_constrained ensures equal size fragments
result = partitioner.partition(molecules)

print(f"Partitioned into {result.n_fragments} fragments.")
for i, frag in enumerate(result.fragments):
    print(f"Fragment {i+1}: {frag.n_molecules} molecules, Charge={frag.molecular_charge}")
```

```python
# Write FMO input
write_gamess_fmo(result.fragments, "water_fmo.inp", fmo_level=2)
print("Written water_fmo.inp")
if os.path.exists("water_fmo.inp"): os.remove("water_fmo.inp")
```
