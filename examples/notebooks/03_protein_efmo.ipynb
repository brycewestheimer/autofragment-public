# Protein EFMO
Fragmentation of Trp-cage (1L2Y) using BioPartitioner.

```python
import sys
sys.path.append('../../src')
from autofragment import BioPartitioner
from autofragment.io import write_gamess_efmo
import os
```

## Partition Protein
We use `BioPartitioner` on the mmCIF file. It handles residue fragmentation automatically.

```python
cif_file = "../data/1l2y.cif"
partitioner = BioPartitioner(infer_bonds=True)
result = partitioner.partition_file(cif_file)

print(f"Created {len(result.fragments)} fragments (residues/waters).")
# Clean names for display
names = [f.id.split('|')[-1] for f in result.fragments[:5]]
print(f"First 5 fragments: {names}")
print(f"Total bonds inferred: {len(result.interfragment_bonds)}")
```

```python
# Write EFMO input
# EFMO requires specific formatting
write_gamess_efmo(result.fragments, "trpcage_efmo.inp", basis="6-31G*")
print("Written trpcage_efmo.inp")
if os.path.exists("trpcage_efmo.inp"): os.remove("trpcage_efmo.inp")
```
