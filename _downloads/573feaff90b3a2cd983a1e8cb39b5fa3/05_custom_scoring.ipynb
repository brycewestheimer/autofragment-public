# Custom Scoring
Implementing a custom partitioner to demonstrate extensibility.

```python
import sys
sys.path.append('../../src')
from autofragment.partitioners.base import BasePartitioner
from autofragment.core.types import Fragment, FragmentTree
from autofragment.io import read_xyz
import numpy as np
```

```python
class ArbitrarySplitPartitioner(BasePartitioner):
    def partition(self, molecules, source_file=None):
        # Simply split list in half regardless of geometry
        mid = len(molecules) // 2
        
        f1 = Fragment.from_molecules(molecules[:mid], "Half1")
        f2 = Fragment.from_molecules(molecules[mid:], "Half2")
        
        source = {}
        if source_file: source = {"file": source_file}
        
        return FragmentTree([f1, f2], source=source)

# Test on aspirin
xyz_file = "../data/aspirin.xyz"
atoms = read_xyz(xyz_file, atoms_per_molecule=1) # Treated as units
part = ArbitrarySplitPartitioner()
res = part.partition(atoms, source_file=xyz_file)

print(f"Custom partition result: {len(res.fragments)} fragments")
print(f"Sizes: {[f.n_atoms for f in res.fragments]}")
```
