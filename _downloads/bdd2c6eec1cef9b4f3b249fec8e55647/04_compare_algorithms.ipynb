# Compare Algorithms
Comparing different partitioning algorithms (kmeans, kmedoids, etc.) on water clusters.

```python
import sys
sys.path.append('../../src')
from autofragment import MolecularPartitioner
from autofragment.io import read_xyz
import time
```

```python
xyz_file = "../data/water_64.xyz"
molecules = read_xyz(xyz_file)
methods = ["kmeans", "kmeans_constrained"]

for method in methods:
    start = time.time()
    p = MolecularPartitioner(n_fragments=4, method=method, strict_balanced=False)
    res = p.partition(molecules)
    dur = time.time() - start
    print(f"Method {method}: {dur:.4f}s, Fragments: {[f.n_molecules for f in res.fragments]}")
```
